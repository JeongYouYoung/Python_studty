```python
# êµ¬ë³„ë¡œ ë°ì´í„°ë¥¼ ì •ë¦¬í•˜ë‹¤.
crime_anal = pd.pivot_table(crime_anal_police.iloc[:,1:], index='êµ¬ë³„', aggfunc=np.sum)
```

```python
crime_anal.head()
```

```python
# ê°•ê°„ê²€ê±°ìœ¨, ê°•ë„ê²€ê±°ìœ¨, ì‚´ì¸ê²€ê±°ìœ¨, ì ˆë„ê²€ê±°ìœ¨, í­ë ¥ê²€ê±°ìœ¨ ì»¬ëŸ¼ ê°€í•˜ê¸°
crime_anal['ê°•ê°„ê²€ê±°ìœ¨'] = crime_anal['ê°•ê°„ ê²€ê±°'] / crime_anal['ê°•ê°„ ë°œìƒ'] * 100
crime_anal['ê°•ë„ê²€ê±°ìœ¨'] = crime_anal['ê°•ë„ ê²€ê±°'] / crime_anal['ê°•ë„ ë°œìƒ'] * 100
crime_anal['ì‚´ì¸ê²€ê±°ìœ¨'] = crime_anal['ì‚´ì¸ ê²€ê±°'] / crime_anal['ì‚´ì¸ ë°œìƒ'] * 100
crime_anal['ì ˆë„ê²€ê±°ìœ¨'] = crime_anal['ì ˆë„ ê²€ê±°'] / crime_anal['ì ˆë„ ë°œìƒ'] * 100
crime_anal['í­ë ¥ê²€ê±°ìœ¨'] = crime_anal['í­ë ¥ ê²€ê±°'] / crime_anal['í­ë ¥ ë°œìƒ'] * 100

# ì‚¬ê±´ì€ ì „ë…„ë„ í•´ê²°ì€ ì´ë²ˆë…„ë„ë¼ì„œ 100 ë„˜ëŠ”ê³³ì´ ìˆë‹¤.
crime_anal.head()
```

```python
# ê²€ê±° ê±´ìˆ˜ëŠ” ì‚­ì œ
del crime_anal['ê°•ê°„ ê²€ê±°']
del crime_anal['ê°•ë„ ê²€ê±°']
del crime_anal['ì‚´ì¸ ê²€ê±°']
del crime_anal['ì ˆë„ ê²€ê±°']
del crime_anal['í­ë ¥ ê²€ê±°']
crime_anal.head()
```

> ì „ë…„ë„ ë°œìƒí•œ ì‚¬ê±´ì˜ ê²€ê±° ê±´ìˆ˜ë„ í¬í•¨ë˜ì–´ì„œ ê²€ê±°ìœ¨ì´ 100ì´ ë„˜ëŠ” ê³³ì´ ë°œìƒí•œë‹¤.

```python
#1) ê²€ê±°ìœ¨ì´ 100ì´ ë„˜ëŠ” ê²½ìš°ì—ëŠ” 100ìœ¼ë¡œ ë³€ê²½í•œë‹¤ 
crime_anal.loc[crime_anal['ê°•ê°„ê²€ê±°ìœ¨'] > 100, ['ê°•ê°„ê²€ê±°ìœ¨']] = 100
crime_anal.loc[crime_anal['ê°•ë„ê²€ê±°ìœ¨'] > 100, ['ê°•ë„ê²€ê±°ìœ¨']] = 100
crime_anal.loc[crime_anal['ì‚´ì¸ê²€ê±°ìœ¨'] > 100, ['ì‚´ì¸ê²€ê±°ìœ¨']] = 100
crime_anal.loc[crime_anal['ì ˆë„ê²€ê±°ìœ¨'] > 100, ['ì ˆë„ê²€ê±°ìœ¨']] = 100
crime_anal.loc[crime_anal['í­ë ¥ê²€ê±°ìœ¨'] > 100, ['í­ë ¥ê²€ê±°ìœ¨']] = 100
crime_anal.head()
```

```python
#2) ê²€ê±°ìœ¨ì´ 100ì´ ë„˜ëŠ” ê²½ìš°ì—ëŠ” 100ìœ¼ë¡œ ë³€ê²½í•œë‹¤ (forë¬¸ ì‚¬ìš©)
# ì§€ê¸ˆì€ ì»¬ëŸ¼ì´ ë¶™ì–´ìˆìœ¼ë‚˜ ë”°ë¡œë”°ë¡œ ë–¨ì–´ì ¸ìˆìœ¼ë©´ ë¦¬ìŠ¤íŠ¸ë¡œ ë§Œë“¤ì–´ì„œ í•´ì£¼ëŠ”ê²Œ ì¢‹ìŒ
updateList = ['ê°•ê°„ê²€ê±°ìœ¨','ê°•ë„ê²€ê±°ìœ¨','ì‚´ì¸ê²€ê±°ìœ¨','ì ˆë„ê²€ê±°ìœ¨','í­ë ¥ê²€ê±°ìœ¨']

for upd in updateList:
    # print(upd)
    crime_anal.loc[crime_anal[upd] > 100, [upd]] = 100

crime_anal.head()
```
