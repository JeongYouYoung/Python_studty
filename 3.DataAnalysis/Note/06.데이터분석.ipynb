# 특성공학( 컬럼추가하기 )

#### FamilySize라는 컬럼을 생성 : SibSp + Parch + 1

```python
# FamilySize라는 컬럼을 생성 : SibSp + Parch + 1
train.head()
train['FamilySize'] = (train.SibSp + train.Parch) +1
train[['SibSp','Parch','FamilySize']].head()
```

#### Embarked 국적별로 분석해볼수 있으니까 S 영국 C 오스트리아 Q 영국 오스트리아에 있나 없나 True . False 컬럼 추가

```python
# Embarked 국적별로 분석해볼수 있으니까 S 영국 C 오스트리아 Q 영국 오스트리아에 있나 없나 True . False 컬럼 추가
train['Nationality_AT'] = train.Embarked == 'C'
train.head()
```

```python
train['Nationality_UK'] =train.Embarked != 'C'
train.head()
```

```python
train[['Embarked','Nationality_AT','Nationality_UK']].head()
```

#### 국적 컬럼 추가하기

- Nationality_AT == False => England
- Nationality_AT == True => Austria

```python
# 국적 컬럼 추가하기
train.loc[train["Nationality_AT"] == False, "Nationality"] = 'England'
train.loc[train["Nationality_AT"] == True, "Nationality"] = 'Austria'
train.head()
```

#### LowFare 15 이하, MediumFare 15~30 이하, HighFare 30이상 (분위값 올림, 내림해서 숫자 맞춘거,)

```python
# Fare
train['Fare'].describe()
```

```python
train["LowFare"] = (train.Fare <= 15 )
train["MediumFare"] = (train.Fare > 15) & (train.Fare <= 30)
train["HighFare"] = (train.Fare > 30 )
train.iloc[:,[8,15,16,17]].head(10)
```

```python
# FaraType 추가하기
train.loc[train["LowFare"] == True, "FareType"] = 'Low'
train.loc[train["MediumFare"] == True, "FareType"] = 'Medium'
train.loc[train["HighFare"] == True, "FareType"] = 'High'
train.iloc[:,[8,18]].head(10)
```

```python
# 함수로 만들어서 써보기
# fare 는 파라미터값이라 데이터 값이 들어오는거 변수이름이 들어오는게 아님.
def check_fare_type(fare):
    if fare < 15:
        return "Low"
    elif fare >= 15 and fare < 30:
        return "Medium"
    else:
        return "High"
```

```python
# 내가 쓴 함수를 불러오겠다. apply()
train['FareType2'] = train['Fare'].apply(check_fare_type)
train[["Fare","FareType2"]].head(10)
```

```python
# Age의 NaN의 평균값으로 대체하기 (평균나이는 소수점 1자리)
AgeMean = round(train.Age.mean(),1)
print(len(train[train["Age"].isnull()]))
train.loc[train["Age"].isnull() , 'Age'] = AgeMean
print(len(train[train["Age"].isnull()]))
```

```python
# 현재까지의 자료 저장하기!
#df.to_excel("../Data/geocoding.xlsx", sheet_name='geocoding_1')
train = train.iloc[:,:19]
train.to_csv("../Data/Titanic.csv")
```
