```python
# NLTK(Natural Language Toolkit) : 자연어 처리 패키지
#불용어 : stopwords, 위에는 영어쪽 단어들
# 언어학에서 사용하는 용어들 : pickle
# corpus : 말뭉치(corpus)는 자연어 분석 작업을 위해 만든 샘플 문서 집합
import nltk
import pickle
from nltk.corpus import stopwords
import re

# 경로 알려주니까 사용 안하면 지우고 나중에 사용할때 다시 설치해도 됨.
nltk.download('all')
```

## text를 불러 word분류

```python
content = open("../Data/Independence.txt").read()
content
```

```python
# 불필요한 심벌 없애기(! ,"", ...)
# cleaned_content = content.replace('!','').replace(",",'') -> 이렇게 할수 있으나 다른방법
cleaned_content = re.sub(r'[^\.\?\!\w\d\s]','', content)
cleaned_content
```

```python
# 대문자를 소문자로 변환
cleaned_content = cleaned_content.lower()
cleaned_content
```

```python
# 문장에서 워드를 토큰으로 쪼갠다.
# 토큰 : 단어 하나하나를 부르는 이름
word_tokens = nltk.word_tokenize(cleaned_content)
print(word_tokens)
```

```python
# 품사분리 -> 튜플로 나누어져 있다. (명사(NN)가지고 주로 사용함)
token_pos = nltk.pos_tag(word_tokens)
print(token_pos)
```

```python
# 명사만 추출하기
# token_pos에서 word, pos가 있는데 pos에 NN 이 있으면 word를어펜드하자
NN_words = []
for word, pos in token_pos:
    if 'NN' in pos:
        NN_words.append(word)

print(NN_words)
```

```python
# 원형으로 복원-> 복수형들 같은걸 원형으로 바꾸자
wlem = nltk.WordNetLemmatizer()
lenmatized_word = []
for word in NN_words:
    new_word = wlem.lemmatize(word)
    lenmatized_word.append(new_word)

print(lenmatized_word)
```

```python
# 불용어 제거 (카운트에 도움 안되는 단어들) => **set을 쓰면 중복된값을 제거하려고 쓰는것.
stopwords_list = stopwords.words('english')
unique_NN_words = set(lenmatized_word)

# 중복된값 제거 하려고 만들어줌
final_NN_words = lenmatized_word

for word in unique_NN_words:
    if word in stopwords_list:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 원하는 단어를 제거
# 불용어 제거는 나중에 전체 다 확인후에 필요 없는 단어들이 보여서 작업하는 단계이다.
customized_stopwords = ['be','today','yesterday',"it's","don't",'others']

unique_NN_words1 = set(final_NN_words)
for word in unique_NN_words1:
    if word in customized_stopwords:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 빈도 분석
from collections import Counter

# Dictionary로 나옴
c=Counter(final_NN_words)
print(c)
```

```python
print(c.most_common(20))
```

```python
# 워드 클라우드(3차원 그래프) => 딕셔너리 이런거 몰라서 하나의 문자열로 만들어주는 작업
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wordcloud

noun_text = ''
for word in final_NN_words:
    noun_text = noun_text + ' ' + word
print(noun_text)
```

```python
# instance 만들기
wordcloud = WordCloud(
    max_font_size=60,
    relative_scaling=0.5

).generate(noun_text) 
# generate가 noun_text의 string value을 입력받아 count 함.
# generate는 데이터를 주지 않아서 count(c)를 쓸줄 알아야함.

plt.figure(figsize=(12, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
