```python
# 5번의 커피를 판매할 수 있는 커피 자판기가 있다.
# 커피의 금액은 300원이다.
# 300원 보다 큰 금액을 입력하면 커피를 주고 (커피 판매횟수 감소) 거스름돈을 준다.
# 커피의 판매 횟수를 소진하면 판매를 중지 한다.
coffee = 5
while coffee != 0:
    price = int(input())
    if price >= 300:
        coffee = coffee -1
        print(f"입금하신 금액은{price} 이고 거스름돈은 {price - 300}원 입니다. \n 잔여 커피의 갯수는 {coffee}입니다.")
        if coffee == 0:
            print("Sold Out")
    else:
        print(f"금액이 부족합니다... \n{300-price}원 만큼 더 넣어주세요") 
```

```python
coffee = 5
while coffee != 0:
  print(f"""
    -----------------------------
    커피 자판기 입니다.
    300원 이상의 금액을 입력하세요 : 
    남은 구매 횟수는 {coffee}번 입니다.
    -----------------------------
    """)
  price = int(input())
  if price >= 300:
    print(f"커피가 나왔습니다. \n 거스름돈은 {price - 300}원입니다.")
    coffee -= 1
    if coffee == 0:
      print("판매를 종료합니다")
  else:
    print("300원보다 큰 금액을 입력해주세요")
  
```
