# 인터넷의 빅데이터
- (빅데이터를 만드는것 : scrapy)
- 빅데이터는 이름 그대로 대규모의 데이터 집합을 의미  
- 데이터를 수집하는 것만으로는 어떤 의미도 없음.  
- 수집한 데이터를 활용해야 가치가 생성.  
### <i>빅데이터라는 용어는 수집한 데이터를 분석해서 비즈니스에 활용하는 것까지를 의미함</i>  
  
    
- 빅데이터를 분석한다는 것은 수많은 데이터에서 규칙성을 찾는 것을 말함.
- 데이터에 포함되어 있는 요소 중에서 중요한 것을을 추출하고 이를 분류해서 규칙성을 찾는 것이 일반적이다.  


### 블로그와 SNS
- 트랜드 분석  
  
### 인터넷 전자 상거래
- 상품 데이터 베이스   
- 네이버와 다음 같은 사이트에서 다양한곳의 상품 데이터를 모아 웹API로 제공  
  
### 금융정보  
- 환율, 주식등의 다양한 금융정보  
- 각 국가의 환율, 주식, 금값등을 실시간으로 추출 가능.  
- 이러한 정보를 정기적으로 추출해서 저장해두고 활용하면 예측등에 활용 할 수 있음.  
  
### 이미지 데이터  
- 플리커, 인스타그램같은 사이트에서 이미지 서비스를 활용하여 이미지를 획득 가능  
- 이미지와 함께 제공되는 태그 정보등을 활용  
  
### 행정기관 정보  
- 공개 데이터 (서울 열린 데이터 광장)  
  
### 위키  
- 고전소설 및 고전 그림  
  
### 머신러링 데이터  
- 공개 데이터 (손글씨 이미지, 사람얼굴 데이터, 동물 데이터 등)

  # 스크레이핑, 크롤링, 데이터 가공  

  
### 스크레이핑 (Scraping)  
- 웹사이트에 있는 특정 정보를 추출하는 기술  
- 웹에 공개된 정보는 HTML형식  
- 요즘 개발자들은 스크레이핑 기본으로 할줄 알아야한다.  
- HTML 데이터를 획득하여 데이터베이스에 저장하려면 가공 기술이 필요  
- 광고 등의 불필요한 정보를 제고하고 필요한 정보를 가져오려면 사이트의 구조를 분석 가능해야함.  
   
### 크롤링(Crawling)  
- 프로그램이 웹사이트를 정기적으로 정보를 추출하는 기술   
   
### 머신러닝이 사용할 수 있는 데이터의 구조  
- 수집한 데이터는 머신러닝을 사용해 다양하게 활용 가능함.  
- 웹에서 내려받은 HTML데이터는 머신러닝이 바로 사용 불가.  
- 이를 해결하기 위해 데이터의 구조를 분석하고 필요한 부분만 추출하는 과정이 필요.  
- 텍스트와 이미지 파일을 아무런 처리없이 학습용 데이터로 사용할 수 없다.  
- 데이터에 어떤 특징이 있는지 **분석가가 직접 확인**하고 가공해야 학습을 시킬 수 있다.
