## requests
#### 응답코드
- 1xx(100번대) : 요청을 받았고 작업 진행중.  
- 2xx(200번대) : 사용자의 요청이 성고적으로 수행됨  
- 3xx(300번대) : 요청은 완료되었으나 리다이렉션(다른 서버이동) 필요.  
- 4xx(400번대) : 사용자의 요청이 잘못됨.  
- 5xx(500번대) : 서버에 오류가 발생함.  

```python
# request 응답 코드
import requests

URL = "https://www.naver.com"
response = requests.get(URL)
print(response.status_code)

#200번 연결 잘되었다.
```

```python
print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기(Python에 관련된걸 검색.)
import requests
URL = "https://search.naver.com/search.naver" #spring
params = {'query' : "python"}
response = requests.get(URL, params=params)
print(response.status_code)
print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기(플러터 관련된걸 검색.)
import requests
URL = "https://search.naver.com/search.naver" #spring
params = {'query' : "플러터"}
response = requests.get(URL, params=params)
print(response.status_code)
print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기
import requests
URL = "https://comic.naver.com/webtoon/detail.nhn" #spring
params = {'titleId' : '68791','no':'157'}
response = requests.get(URL, params=params)
print(response.status_code)
print(response.text)
```
