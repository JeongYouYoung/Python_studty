- https://movie.daum.net/ranking/boxoffice/weekly

#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://movie.daum.net/ranking/boxoffice/weekly"
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
li_list = soup.select("#mainContent > div > div.box_boxoffice > ol > li > div > div.thumb_cont > strong > a")
li_list
```

```python
# class(class="link_txt") 가 보였으니까 이런 방식으로 사용가능
li_list = soup.select("#mainContent .link_txt")
li_list
```

```python
count = 0
titleLists = [] # 영화제목 Lists
countLists = [] # 영화 순위 Lists
for list in li_list:
    count += 1
    #strip() 공간 띄어쓰기 제거 strip() 잘쓰기!
    print(count, list.string.strip())
    # data.frame 만들기
    countLists.append(count)
    titleLists.append(list.string.strip())
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countLists 만드로 DataFrame 생성
# Column 만들기
df = pd.DataFrame(data = countLists, columns=['순위'])
df
```

```python
# 기존 DataFrame에 컬럼과 Lists추가
# Python은 벡터값을 Series라고 부름
df['영화제목'] = pd.Series(titleLists)
df
```

```python
# csv로 저장
# index = False 는 앞에 row번호 있는 부분!
df.to_csv("../Data/DaumList.csv", encoding='utf-8', index = False)
```

# 네이버 랭킹 사이

- https://movie.naver.com/movie/sdb/rank/rmovie.naver

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
#old_content > table > tbody > tr:nth-child(2) > td.title > div > a
naver_list = soup.select("td.title > div > a")
naver_list[0]["title"]
```

```python
count = 0
n_titleLists = [] # 영화제목 Lists
n_countLists = [] # 영화 순위 Lists
for list in naver_list:
    count += 1
    #strip() 공간 띄어쓰기 제거 strip() 잘쓰기!
    print(count, list.string.strip())
    # data.frame 만들기
    n_countLists.append(count)
    n_titleLists.append(list.string.strip())

# Lists 확인
print(n_countLists)
print(n_titleLists)
```

```python
# countLists 만드로 DataFrame 생성
# Column 만들기
n_df = pd.DataFrame(data = n_countLists, columns=['index'])
n_df
```

```python
# 기존 DataFrame에 컬럼과 Lists추가
# Python은 벡터값을 Series라고 부름
n_df['Title'] = pd.Series(n_titleLists)
n_df
```

```python
# csv로 저장
# index = False 는 앞에 row번호 있는 부분!
n_df.to_csv("../Data/naverRanking.csv", encoding='utf-8', index = False)
```
