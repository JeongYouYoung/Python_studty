# 머신러닝 이란?
> 수 많은 데이터를 학습시켜 거기에 있는 패턴(반복,y=ax+b)을 찾아내고 이를 기반으로 데이터를 분류하거나 예측.    
- (랜덤, 컴퓨터의 엄청난 반복작업 이두개가 머신러닝을 만듦, 알고리즘 : 규칙 만드는것-> 알고리즘 => y=ax+b 이 패턴을 찾아주는거)  
- 알고리즘은 : y=ax+b인데  a값과 b의 값을 찾아주는것(y^)이 머신러닝  
- 일반 소프트웨어(규칙(공식)이 정해져있어, 알고리즘을 사람이 정함)[결정론적 소프트웨어] 는 언제나 똑같은 과정을 거쳐 똑같은 결과를 나타냄.   
- 머신러닝(비결정론적 소프트웨어) 기반 소프트웨어는 데이터와 알고리즘에 따라 때로는 다른 결과를 추론해 냅니다.  
- 추론 : 이미 알고있는 것으로부터 논리적 결론을 도출하는 과정  (y^ : 근사값)

<br>

## 결정론적 소프트웨어와 비결정론적 소프트웨어의 차이점
예를 들어 지문인식기의 소프트웨어를 결정론적 소프트웨어로 처리하였다고 하면 -> 지문입력기에 지문이 조금만 틀어져도 거부 메세지가 나온다.  
이에 반해 비결정론적 소트프웨어는 지문이 조금 틀어져도 추론을 통해 지문의 가부를 결정함.

<br>
<br>

## 머신러닝 개요
> 머신러닝이랑 인공지능 연구 과제중의 하나로 인간의 뇌가 자연스럽게 수행하는 '학습'이라는 능력을 컴퓨터로 구현하는 방법   
- 어느정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석 시킵니다.
- 그리고 이때 데이터애소 일정한 규칙을 찾아 냅니다.  
- 그리고 찾아낸 규칙을 기반으로 다른 데이터를 분류(classification)하거나 예측(regression)하는 것.  
<br>
- 과적합 :  훈련데이터는 잘맞추는데 실전에서 쓸수 없어   

<br>
<br>

## 머신러닝의 종류

#### 지도학습(Supervised Leaning)
> 학습시 데이터와 답을 같이 학습한후 다른 데이터의 값을 예측  
- 데이터가 많아야함   

#### 비지도학습(Unsupervised Leaning)
> 학습시 데이터만 학습한후 다른 데이터의 규칙성을 찾습니다(답이 없어). 본질적인 구조등을 확인할때 사용

#### 강화학습(Reinforcement Leaning)
> 부분적으로 답을 입력한후 데이터를 기반으로 최적의 답을 찾아 냅니다.  
- 답을 알려주지만 지도학습이랑 서로 다름. 수많은 시행착오를 걸쳐서 답을 찾아내고 또다른 시행착오를 거쳐 답을 찾아냄  
(ex: 알파고=딥러닝 + 강화학습)  

#### 딥러닝(Deep Leaning)
> 지도학습과 비지도학습의 관련이 있으나 특이성은 학습자료에 대한 인간의 개입을 받지 않는다.    
- (난이도 : 딥러닝 < 강화학습 < 비지도학습 < 지도학습)

### 머신러닝의 흐름
(1) 데이터 수집  
(2) 데이터 가공  
(3) 데이터 학습  
&nbsp;&nbsp;&nbsp;(3-1) 학습 방법 선택  
&nbsp;&nbsp;&nbsp;(3-2) 매개 변수 조정  
&nbsp;&nbsp;&nbsp;(3-2) 모델 학습  
(4) 모델 평가  
(5) 예측률이 좋은 경우 다음 단계 진행하고 아니면 (3)번 부터 다시  
(6) 성공

# 머신러닝 응용 분야
### 클래스 분류(Classification)
- 특정 데이터에 레이블(label: 정답값,범주형 데이터값)을 붙여 분류할 수 있습니다.
- 스팸 메일 분류, 필기인식, 증권 사기 방지 등
- 타이타닉의 생존과 사망, 아이리스의 세토사 버시칼라 등 나누는것


### 클러스터링(Clustering) : 그룹 나누기
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있습니다.
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공


### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터를 추천하는 것 입니다.
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천.


### 회귀(Regression)
- 과거의 데이터를 기반으로 미래를 예측한 것 입ㄴ디ㅏ.
- 판매 예측, 주가 변동


### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것 입니다.
- 데이터를 시각화 하고나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약할때 사용(퍼포먼스를 위해)

# 프로젝트 과정
- 머신러닝 프로젝트는 크게 문제정의, 데이터 획득, 모델 구현, 검증, 테스트의 단계로 구현

### 문제정의
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계
- 대략적으로 어떤 데이터를 수집해서 어떤 머신러닝 알고리즘으로 문제를 해결할지 결정하는 단계

### 데이터 획득
- 머신러닝 모델 학습에 필요한 데이터를 수집하는 단계
- 성공적인 머신러닝 프로젝트는 뛰어난 머신러닝 실력보다도 얼마나 충분히 데이터를 획득했느냐에 의해 결정됩니다.
- 획득한 데이터는 학습데이터, 검증데이터, 테스트데이터로 구분 되어 사용됩니다.
- 전체데이터의 20% 는 일반화 성능을 측정하기위한 테스트 데이터로 사용 80%중 20%는 검증데이터로 나눈다, 60%로 학습을 시킴.

### 모델 구현
- '학습 데이터'를 기반으로한 한개이상의 머신러닝 모델을 구현

### 검증
- 검증 데이터를 사용해 구현된 머신러닝 모델들의 성능을 검증
- 최종적으로 검증 결과가 가장 높은 모델 하나를 선택

### 테스트
- 검증 결과를 통해 선택된 모델의 성능을 테스트 데이터로 측정한다.(일반화 성능 측정 => 과소적합과,과대적합 판단하기위해)
