## Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
# 1번째 Hidden Layer : 256 노드 (2의 배수, 임의의 갯수(Hyper Parameter))
# activation = sigmoid, tanh, relu, softmax, leakyRelu
model.add(keras.layers.Dense(256, activation='tanh', input_shape=(3,))) # 은닉층
model.add(keras.layers.Dense(1, activation='sigmoid')) # 출력층
```

Optimizer => Learning rate를 알맞게 조정해줌 ->(y - y^)
- adam
- adagrad
- adadelta
- rmsprop
- sgd  

loss : 손실함수
- binary_crossentropy : 분류 및 확률 예측 함수

metrics : 모델의 평가 기준


```python
# 손실함수
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

```python
# 학습 시키기
# 일반적인 fit은 못씀 pandas에 들어있어서 -> numpy로 바꿔줘야해
import numpy as np
model.fit(np.array(x), np.array(y), epochs=499)
```

```python
# 예측하기
model.predict(
    [
        [700, 3.7, 1],
        [700, 3.7, 2],
        [700, 3.7, 3],
        [700, 3.7, 4]
    ]
)
```
